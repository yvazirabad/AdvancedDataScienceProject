```python
df[['EVENT_TYPE', 'CZ_NAME']] = df[['EVENT_TYPE', 'CZ_NAME']].astype('str')
df[['INJURIES_DIRECT', 'INJURIES_INDIRECT', 'DEATHS_DIRECT', 'DEATHS_INDIRECT']] = df[['INJURIES_DIRECT', 'INJURIES_INDIRECT', 'DEATHS_DIRECT', 'DEATHS_INDIRECT']].astype('int64')
df[['BEGIN_LAT','BEGIN_LON']] = df[['BEGIN_LAT','BEGIN_LON']].astype('float')
#df[['BEGIN_DATE_TIME']] = df[['BEGIN_DATE_TIME']].to_datetime()
print(df.dtypes)
```

```python
# This code block makes me wish I had 64 GB of ram......
for col in ['DAMAGE_PROPERTY', 'DAMAGE_CROPS']:
    for index, value in enumerate(df[col]):
        if 'K' in str(value):
            df.loc[index,col] = value.replace('K','')
df[['DAMAGE_PROPERTY', 'DAMAGE_CROPS']] = df[['DAMAGE_PROPERTY', 'DAMAGE_CROPS']].astype('int64')
#for col in ['DAMAGE_PROPERTY', 'DAMAGE_CROPS']:
#    for index, value in enumerate(df[col]):
#       df[col][index] = value*1000
print(df.types)
```
